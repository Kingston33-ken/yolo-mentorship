```python
from ultralytics import YOLO

# Load the AI model
model = YOLO("yolov8n.pt")

# Run detection on your uploaded image
results = model("Mike and El.jpeg")

# Print the results in a clean list
print("\n========== DETECTION RESULTS ==========")
for result in results:
    boxes = result.boxes
    if len(boxes) == 0:
        print("No objects were detected.")
    else:
        print(f"Number of objects detected: {len(boxes)}\n")
        for i, box in enumerate(boxes):
            object_name = result.names[int(box.cls)]
            confidence_percent = float(box.conf) * 100
            print(f"Object {i+1}: {object_name} ({confidence_percent:.1f}%)")
print("========================================")
```
